```python
# First Phyton Program by Onur Baser

from pathlib import Path
import csv
# Set the file path
csvpath = Path("budget_data.csv")

# Initialize dictionary
analysis = {}
difference={}
change={}
maxchange={}

# Initialize profit_losses
profit_losses = []
totdifference=[]
 

# Initialize line_num variable
line_num = 0
j=1
i=1

total_profit_loss=0
count_profit_loss=0
 

# Open the csv file as an object
with open(csvpath, "r") as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)

   
# Read each row of data after the header
    for row in csvreader:
        month_year = row[0]
        profit_loss = int(row[1])
        profit_losses.append(profit_loss)
        analysis[line_num]= {"month_year":month_year,"profit_loss":profit_loss}
        change[line_num]={profit_loss}
        line_num+=1
        
        
    # Calculating total months and total amount
    
    for profit_loss in  profit_losses:
    # Sum the total and count variables
        total_profit_loss += profit_loss
        count_profit_loss += 1
        
     
  # Calculating the change at each row     
    while i < len(analysis):
        change[i] = analysis[i]["profit_loss"]-analysis[i-1]["profit_loss"]
        i+=1
        
  #Assigned first value=0 and find min and maximum key numbers in change. 
 
    change[0]=0
    max_key=max(change, key=change.get)
    min_key=min(change, key=change.get)
    
    
  # Calcuation average change  
    total=0
    i=1
    while i < len(analysis):
        total= total + change[i] 
        i+=1
    average=round(total/(len(analysis)-1),2)
         
        
    #To read month and year, created dictinary with two values at each key
    

    while j < len(analysis):
        differ  = analysis[j]["profit_loss"]-analysis[j-1]["profit_loss"]
        difference[j]={"month_year":analysis[j]["month_year"], "monthly_difference":differ}
        j+=1

   #Derive requested variables
        
    month_inc_prof= difference[max_key]["month_year"]
    max_prof_value= difference[max_key]["monthly_difference"]
    month_dec_prof= difference[min_key]["month_year"]
    max_loss_value= difference[min_key]["monthly_difference"]
    
    #Printing to the terminal
    
    print ("Financial Analysis")  
    print ("--------------------------------------")
    print(f"Total Months: {count_profit_loss}")
    print(f"Total: ${total_profit_loss}")
    print(f"Average Change: ${average}") 
    print(f"Greatest Increase in Profits: {month_inc_prof} (${max_prof_value})")
    print(f"Greatest Decrease in Profits: {month_dec_prof} (${max_loss_value})")
 


# Set the output header
header = ["Total Months", "Total", "Average Change", "Month_Year of Greatest Increase in Profit", "Greatest Increase in Profit","Month_Year of Greatest Increase in Profit", "Month_Year of Greatest  Decrease in Profit" ]
# Create a list of metrics
metrics = [count_profit_loss, total_profit_loss, average,month_inc_prof,max_prof_value,month_dec_prof, max_loss_value]


   # Set the output file path
output_path = Path('output.csv')

# Open the output path as a file object
with open(output_path, 'w') as csvfile:
    # Set the file object as a csvwriter object
    csvwriter = csv.writer(csvfile, delimiter=',')
    # Write the header to the output file
    csvwriter.writerow(header)
    # Write the list of metrics to the output file
    csvwriter.writerow(metrics)


    
```
